**Q3: Create a markdown cell and explain the meaning of each column**

In jupyter notebook it can be done selecting "Markdown" instead of "Code" from "Menu bar" (have doubt what exacty it is called), or using shortcut. In Google colab, same purpose can be served using the "Add text cells". 

Below is the short description of each column.

Player: Name of the Player followed by the country/group/associations they represent in their career which are counted as international matches and for this record.

Span: Time represented as "starting year" to "ending year" between which the mathces have been played. (Q: what if a player is currently playing? to check this the original data source has been visited, and found that, in that case, the current year has been mentioned, but no missing "end year". It is easier in that case as the data set is small and the players playing currently is known, but in case of very big, unorganized data set with little/poor/incorrect description, it might need to take different appraoch to find it.)

Mat: short form of "Matches", means the number of game have been played. (Q; what if a match had been abandoned without bowling a single bowl? No answer found.)

Inns: means "Innings bowled", most probably, if the bowler bowl at least a single bowl in a match is counted as "Inns" (so the question arises in last point is solved here.)

Balls: Number of valid delivery completed by the bowler (excluding all no balls, wide balls, dead balls and other not counted as legal delivery).

Runs: Runs conceded by the bowler (scored by a batsman or awarded as extra like wide). Remember an illegal delivery can cost some runs also.

Wkts: Wicket taken by the bowler (example, run out is not included here as this is not counted as "taken by the bowler").

BBI: Best bowling figure in a single innings. This is sorted accoridng to the highest number of wicket taken, then least run conceded, then number of legal delivery bowled, respectively.

Ave: Bowling average, means the total runs conceded is divided by the total wicket taken in total career (Total Runs Conceded (Column "Runs") / Total number of wickets taken (Column "Wkts")).

Econ: Economy rate, which is defined as the run conceded per over (1 over == 6 balls) according to total career data (Total run conceded (Column "Runs")/ Total number of valid delivery (Column "Balls") * 6).

SR: bowling strike rate, means number of balls needed by this bowler to ball to take a wicket (Total number of ball bowled/Total number of wicket).

4: Four wickets haul, means number of innings in which the bowler takes 4 or more wickets.

5: Five wickets haul, means number of innings in which the bowler takes 5 or more wickets.

Another way could be rearranging the "Player_name (Country)" column in a way thus it makes the dataset "neat" and extarcting information become easier, in this section, it will be done

```python
#creating two separate column with player name and the country they represent and replacing the original "Player_name (Country)" Column
df[["Player_Name", "Country"]] = df["Player_name (Country)"].str.split("(", expand=True)
df ["Country"] = df ["Country"].str.replace(")", "")
display(df.head())
```

```python
#paste the output of this command and create a new list, inside the list rearrange the column name, much easier than writing down separately
new_sequence_column = ['Player_Name', 'Country', 'Career_Span', 'Matches_Played', 'Innings_Bowled', 'Balls_bowled', 'Runs_Conceded', 'Wickets_Taken', 'Bowling_Average', 'Economy_Rate', 'Strike_Rate', '4_Wicket_haul', '5_Wicket_haul']

#new dataset replace the old dataset
df = df[new_sequence_column]
display(df.head()) #check the purpose is served
```

```python
df['Country'].value_counts() #to see how many unique value in coutnries to plan a separation
```

Plan is to remove anything other than country name from the column country (remember WI is counted as a "Country" in ICC event, not an association, before doing that, separate columns  are created thus the data is not lost any if any onformation regarding represting association other than country is required can be seen

```python
#defining 3 functions separately to find if those association are present in "Country" column or not, in first one conventional method is followed, for next 2 lambda function is used

def asia_check(x):
    if "Asia" in x:
        return "Yes"
    else:
        return "No"

africa_check = lambda x: "Yes" if "Afr" in x else "No"
icc_check = lambda x: "Yes" if "ICC" in x else "No"
```

```python
#now creting 3 new columns to keep the data relataed to represting the associations

df['represents_ICC'] = df['Country'].apply(icc_check)
df['represents_Asia'] = df['Country'].apply(asia_check)
df['represents_Africa'] = df['Country'].apply(africa_check)

display(df.head()) #check does it work
```

```python
#Now try to answer from is new dataset, how many player represent ICC
(df['represents_ICC']=="Yes").sum()
```

Now try to make it "nicer", remove association names from the column "Country"

```python
str_to_remove = ["Asia/", "Afr/", "ICC/", "/ICC",]
new_country_name = []
old_country_name = df["Country"].to_list()

for country in old_country_name:
    for string in str_to_remove:
        country=country.replace(string, "")
    new_country_name.append(country)
print(new_country_name)

    
```

```python
df["Country"] = new_country_name #replace the old column with new list
display(df.head()) #check the dataset is as per requirement
```

**Q10: Which player(s) had played for the longest period of time?**

To answer this question, the career_span column need to be rearranged to find the start year and end year of the career, creating two separate column with these value and converting the data type to int will make it easy to find the duration, then simple sorting will help to get the answer.

```python
# splitting the 'Career_Span' column into two
df[['Start_year', 'End_year']] = df['Career_Span'].str.split("-", expand=True)

# remove old column
df = df.drop("Career_Span", axis=1)

#converting data type of new two columns to int type

df['Start_year'] = df['Start_year'].astype('int') 
df['End_year'] = df['End_year'].astype('int')

#creating new column to find career duration in year
df['Career_duration'] = df['End_year'] - df['Start_year']

display(df.head())
```

```python
#now sort the dataset in descending order of Career_duration and find top 10
df.sort_values(by='Career_duration', ascending = False).head(10)
```
